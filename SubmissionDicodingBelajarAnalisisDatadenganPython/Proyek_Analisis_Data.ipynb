## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana Faktor cuaca dapat mempengaruhi Penyewaan sepeda?

```python
# Data dan warna cuaca dari data per harinya
cuaca_colors = {
    'Clear': '#59D5E0',
    'Cloudy': '#F5DD61',
    'Light Rain': '#FAA300',
    'Heavy Rain': '#F4538A'
}

# Menggabungkan total penyewaan berdasarkan musim dan cuaca
cuaca = day_data.groupby(['Season', 'Weather'])['Total_rental'].sum().reset_index()
spring_data = cuaca[cuaca['Season'] == 'Spring']
summer_data = cuaca[cuaca['Season'] == 'Summer']
fall_data = cuaca[cuaca['Season'] == 'Fall']
winter_data = cuaca[cuaca['Season'] == 'Winter']

# Membuat diagram lingkaran untuk setiap musim
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for ax, season, data in zip(axes, ['Spring', 'Summer', 'Fall', 'Winter'], [spring_data, summer_data, fall_data, winter_data]):
    cuaca_gp = data.groupby('Weather')['Total_rental'].sum().reset_index()
    colors = []
    for weather in cuaca_gp['Weather']:
        try:
            colors.append(cuaca_colors[weather])
        except KeyError:
            # Jika kunci tidak ditemukan dalam kamus, gunakan warna default
            colors.append('gray')
    ax.pie(cuaca_gp['Total_rental'], labels=cuaca_gp['Weather'], autopct='%1.1f%%', startangle=90, colors=colors)
    ax.set_title(f'Proporsi Cuaca pada Musim {season}')

plt.tight_layout()
# Simpan gambar dengan nama "gambar1"
plt.savefig('image/export/gambar1.png')
plt.show()
```

### Pertanyaan 5: Bagaimana perbandingan antara jumlah penyewaan sepeda setiap bulan dalam satu tahun dengan jumlah total penyewaan sepanjang tahun?

```python
# Mengubah tipe data kolom Month menjadi kategori per harinya
day_data['Month'] = pd.Categorical(day_data['Month'], categories=
    ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],
    ordered=True)

# Mengelompokkan data berdasarkan bulan dan tahun
monthly_counts = day_data.groupby(by=["Month","Year"]).agg({
    "Total_rental": "sum"
}).reset_index()

# Memplot jumlah total sepeda yang disewakan berdasarkan Bulan dan Tahun
sns.lineplot(
    data=monthly_counts,
    x="Month",
    y="Total_rental",
    hue="Year",
    palette="rocket",
    marker="o"
)

plt.title("Jumlah Total Sepeda yang Disewakan per Bulan dan Tahun")
plt.xlabel("Bulan")
plt.ylabel("Jumlah Total Sepeda Disewakan")
plt.legend(title="Tahun", loc="upper right")
plt.xticks(rotation=45)  
plt.tight_layout()

# Simpan gambar dengan nama "gambar5"
plt.savefig('image/export/gambar5.png')
plt.show()
```
